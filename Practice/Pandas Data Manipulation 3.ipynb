```python
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/penguins_cleaned.csv')
df.isnull().sum()
```

```python
# Here, we will assign a missing value to it

df.iloc[0, 0] = pd.NA
df.iloc[5, 1] = pd.NA
df.iloc[10, 2] = pd.NA
df.iloc[15, 3] = pd.NA
df.iloc[20, 4] = pd.NA
df.iloc[25, 5] = pd.NA
df.iloc[30, 6] = pd.NA
df.iloc[40, 6] = pd.NA
```

```python
# Individually test each cell whether it has a missing value, it will return True/False

df.isnull()
```

```python
df.isna()
```

```python
# Display only rows with at least 1 missing value
# Note: axis of 0 = rows, axis of 1 = columns

df[df.isnull().any(axis=1)]
```

```python
# Check for missing data by printing the sum of missing values for each column
df.isnull().sum()
```

```python
#Check for Non-Missing Data
df.notna() #or df.notnull()
```

```python
# Drop rows that have at least 1 missing value
df.dropna()
```

```python
# Note that the above did not save to the existing DataFrame
df.dropna(inplace=True)
```

```python
df
```

```python
# Replace missing values with 0
# Not an optimal approach

df.fillna(0)
```

```python
df.species.fillna('-', inplace=True)
```

```python
df
```

```python
# Replace missing values with the column means
# Better, but does not work for categorical values

df.fillna(df.mean())
```

```python
import pandas as pd
import numpy as np


np.random.seed(20)
df = pd.DataFrame(np.random.randn(10,4),index=pd.date_range('2021-07-18',periods=10),columns=['a','b','c','d'])
```

```python
df
```

```python
df['logofD'] =  df['d'].apply(lambda x:np.log(x))
```

```python
df
```

```python
def test(x):
    if(x<0):
        return x
    else:
        np.log(x)
        
df['logD_fun'] = df['d'].apply(test)
```

```python
df
```

```python
def test1(x):
    if x < 0 :
        return 1
    else:
        return 0
    
df['zeroOne'] = df['d'].apply(test1)
```

```python
df
```

```python
df.sort_values(by='zeroOne')
```

```python
df['index_col'] = df.index
```

```python
df
```

```python
df.dtypes
```

```python
df['day'] = df.index.day
df['month'] = df.index.month
df['year'] = df.index.year
df['dayName'] = df.index.day_name()
```
