```python
import statsmodels.api as sm
import pylab as pyl
```

```python
# np.random generates different random numbers
# whenever the code is executed
# Note: When you execute the same code
# the graph look different than shown below.

# Random data points generated
data_points = np.random.normal(0, 1, 100)

sm.qqplot(data_points, line ='45')
pyl.show()
```

```python
logpoints = np.random.lognormal(0,1,100)
logpoints
```

```python
# Log normal distribution 
sns.displot(logpoints,kind='kde')
```

```python
sm.qqplot(logpoints, line ='45') 
pyl.show()
```

```python
transformedPts = np.log(logpoints)
```

```python
# Log normal distribution transformed to Normal Distribution
sm.qqplot(transformedPts, line ='45')
pyl.show()
```
