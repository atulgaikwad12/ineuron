```python
df = pd.DataFrame(np.random.rand(3,2),columns=['a','b'])
df
```

```python
print('Mean\n' , df.mean()) # column wise mean
print('Correlation\n' , df.corr()) # corr of each column with other columns
print('Std\n' , df.std()) # standard distribution of columns
print('Agg\n' , df.agg('a')) # aggregate as per column a
print('Covariance\n' , df.cov())
```
