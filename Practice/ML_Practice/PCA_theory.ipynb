```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('glass.data')
df 
```

```python
df.isna().sum() # no null in case of null value do imputation
```

```python
df_x = df.drop(columns=['index','Class'],axis=1)
df_x
```

```python
from sklearn.preprocessing import StandardScaler
scaler =  StandardScaler()
x_scaled = scaler.fit_transform(df_x)
```

```python
df_scaled = pd.DataFrame(x_scaled, columns=df_x.columns)
df_scaled
```

```python
df_scaled.describe() # now mean is almost 0 and std is approx 1 
```

```python
from sklearn.decomposition import PCA
pca = PCA()
principal_compos = pca.fit_transform(df_scaled)
```

```python
pd.DataFrame(principal_compos)
```

```python
pca.explained_variance_ratio_
```

```python
# Screen Plot 

import matplotlib.pyplot as plt
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("Number of required comoponents")
plt.ylabel("Explained Variance Ratio")

# as per graph to retain meaning of whole data i.e max EVR is near 6 no of comoponents PCA required 
# above 0.9 is fine EVR value
```

```python
#https://drive.google.com/file/d/109azU0trfoeLRDKQpHUybr4bLhurcUMM/view
data = pd.read_csv('D:\ineuron\ineuron\Practice\ML_Practice\datasets\Mall_Customers.csv')
```

```python
data
```

```python
x = data.drop(columns=['CustomerID','Genre','Age'],axis=1)
x
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,15):
    kmean = KMeans(n_clusters=i, init='k-means++', random_state=30)
    kmean.fit(x)
    wcss.append(kmean.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,15), wcss)
# we can select no of cluster between 5-6 
```

```python
kmean1 = KMeans(n_clusters=6, init='k-means++', random_state=30)
kmean1.fit_predict(x)
```

```python
x['cluster_number'] = kmean1.fit_predict(x)
x
```

```python
x[x['cluster_number'] == 4]
```

```python
x[x['cluster_number'] == 1]
```

```python
kmean1.predict([[55,31]])
```

```python
from sklearn.cluster import MiniBatchKMeans
minik = MiniBatchKMeans(random_state=30)
minik_pred_clusters = minik.fit_predict(x.drop(columns=['cluster_number'],axis=1))
minik_pred_clusters
```

```python
minik.predict([[55,31]])
```

```python

from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=1,min_samples=3)
dbscan.fit_predict(x.drop(columns=['cluster_number'],axis=1))
```

```python
predicted_clusters = dbscan.labels_
predicted_clusters
```

```python
set(predicted_clusters)
```

```python
from sklearn import metrics
```

```python
metrics.adjusted_rand_score(x['cluster_number'],minik_pred_clusters) # considering k-mean++ prediction as true value 
# and mini k mean prediction as predicted 
```

```python
metrics.jaccard_score(x['cluster_number'],minik_pred_clusters,average=None)
```

```python
metrics.jaccard_score(x['cluster_number'],predicted_clusters,average=None)
```
